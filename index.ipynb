*PHASE 2: Exploratory Data Analysis (EDA) /Feature Enginnering*


Creating Usage Frequency

Getting value count in Churn column

```python
count = clean_dataset['Churn'].value_counts()
display(count)
```

Do long term customer churned less?


```python
avg_tenure = clean_dataset.groupby('Churn')['tenure'].mean()
print('The average of churned and non churned customer per month is:')
print(avg_tenure)
```

Do expensive plans churn more?

```python
clean_dataset[['Churn','MonthlyCharges']]
```

Do expensive plans churn more?

```python
avg_plan = clean_dataset.groupby("Churn")['MonthlyCharges'].mean()
display(avg_plan)
```

Which contract type has highest churns?


```python
churn_percentage = pd.crosstab(clean_dataset['Contract'],clean_dataset['Churn'],normalize = 'index')*100
print(churn_percentage)
```

Customer Segementation Chart

- HIGH VALUE
- MEDIUM VALUE
- HIGH VALUE


```python
def assign_value(total_charges):
    if total_charges <=401.45:
        return 'Low Value'
    elif total_charges <=3794.74:
        return 'Medium Value'
    else:
        return 'High Value'
clean_dataset['CustomerSegment'] = clean_dataset['TotalCharges'].apply(assign_value)
clean_dataset[['TotalCharges','CustomerSegment']].head(3)
```

```python
segments = clean_dataset['CustomerSegment']
colors = ['Blue' , 'darkblue','lightblue']
segment_counts = segments.value_counts()
plt.bar(segment_counts.index, segment_counts.values, color=colors)
plt.title("Customer Segment Distribution")
plt.xlabel("Customer Segment")
plt.ylabel("Count")
plt.show()
```
