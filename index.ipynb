Customer Segementation Chart

- HIGH VALUE
- MEDIUM VALUE
- HIGH VALUE


```python
def assign_value(total_charges):
    if total_charges <=401.45:
        return 'Low Value'
    elif total_charges <=3794.74:
        return 'Medium Value'
    else:
        return 'High Value'
dataset['TotalCharges'] = pd.to_numeric(dataset['TotalCharges'], errors='coerce')
dataset = dataset.dropna(subset=['TotalCharges'])
dataset['CustomerSegment'] = dataset['TotalCharges'].apply(assign_value)
dataset[['TotalCharges','CustomerSegment']].head(3)
```

```python
segments = dataset['CustomerSegment']
colors = ['Blue' , 'darkblue','lightblue']
segment_counts = segments.value_counts()
plt.bar(segment_counts.index, segment_counts.values, color=colors)
plt.title("Customer Segment Distribution")
plt.xlabel("Customer Segment")
plt.ylabel("Count")
plt.show()
```

Correlation Heatmap - of all numerical columns in dataset

```python
numeric_cols=dataset.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
bins = [0,12,24,72]
labels = ['Month-to-Month','One Year','Two Year']
dataset['tenuregroup']=pd.cut(dataset['tenure'],bins=bins,labels=labels)
plot_data = dataset.groupby(['tenuregroup','Churn']).size().unstack(fill_value = 0)
plot_data.plot(kind='bar',stacked=True,color=['black','#C11C84'])
plt.show()
```
